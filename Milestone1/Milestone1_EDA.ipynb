## Data merging

### Subtask:
Place the cell that merges the two dataframes on the common key.


**Reasoning**:
Identify the common column and merge the two dataframes, then display the shape and head of the merged dataframe.



```python
# Display first few rows to identify common key
print("Customer Details Columns:\n", df1.columns)
print("Basket Details Columns:\n", df2.columns)

# Assuming both have a common column like 'customer_id'
common_key = 'customer_id'

# Merge the two tables
merged_df = pd.merge(df1, df2, on=common_key, how='inner')
print("‚úÖ Merge successful. Shape of combined dataset:", merged_df.shape)
display(merged_df.head())
```

## Data preprocessing

### Subtask:
Organize cells for handling missing values, duplicates, and data type conversions on the merged dataframe.


**Reasoning**:
Check for missing values, fill them with the mean for numeric columns, remove duplicates, convert data types, and print the info of the preprocessed dataframe to verify the changes.



```python
# Check for missing values
print("Missing Values per Column:\n", merged_df.isnull().sum())

# Fill or drop missing values (Filling with mean is only for numeric columns)
numeric_cols = merged_df.select_dtypes(include=np.number).columns
merged_df[numeric_cols] = merged_df[numeric_cols].fillna(merged_df[numeric_cols].mean())

# Remove duplicate rows
merged_df.drop_duplicates(inplace=True)

# Fix incorrect data types
merged_df = merged_df.convert_dtypes()

print("‚úÖ Preprocessing complete.")
merged_df.info()
```

## Descriptive analysis

### Subtask:
Include cells for summary statistics, correlation matrix, and outlier detection on the merged dataframe.


**Reasoning**:
Include cells for summary statistics, correlation matrix, and outlier detection on the merged dataframe.



```python
# Summary statistics
print("Summary Statistics:\n")
display(merged_df.describe())

# Correlation matrix (excluding non-numeric columns)
print("\nCorrelation Matrix:\n")
numeric_merged_df = merged_df.select_dtypes(include=np.number)
display(numeric_merged_df.corr())

# Detect outliers using IQR
Q1 = numeric_merged_df.quantile(0.25)
Q3 = numeric_merged_df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_merged_df < (Q1 - 1.5 * IQR)) | (numeric_merged_df > (Q3 + 1.5 * IQR))).sum()
print("\nOutliers per column:\n", outliers)
```

## Exploratory data analysis (eda)

### Subtask:
Arrange cells for generating visualizations (histograms, boxplots, scatter plots, bar plots, heatmaps, count plots) on the merged dataframe to uncover trends and patterns.


**Reasoning**:
Generate various plots on the merged dataframe to visualize distributions, relationships, and frequencies, which aligns with the subtask of generating visualizations for EDA.



```python
# 1. Histograms for numeric columns
numeric_cols = merged_df.select_dtypes(include=np.number).columns.tolist()
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(merged_df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# 2. Boxplots for numeric columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df.select_dtypes(include=np.number))
plt.title('Boxplot of Numeric Columns')
plt.ylabel('Value')
plt.show()

# 3. Scatter plots for relevant pairs of numeric columns
# Based on the correlation matrix, customer_id and tenure show a strong negative correlation.
# Let's plot customer_id vs tenure and customer_age vs tenure
relevant_pairs = [('customer_id', 'tenure'), ('customer_age', 'tenure')]

for x_col, y_col in relevant_pairs:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=merged_df[x_col], y=merged_df[y_col])
    plt.title(f'Relationship between {x_col} and {y_col}')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()

# 4. Count plots for categorical columns
# Identify categorical columns again after convert_dtypes
categorical_cols = merged_df.select_dtypes(include='string').columns.tolist()
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=merged_df[col])
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

# 5. Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(merged_df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Columns')
plt.show()

# 6. Bar plots for categorical vs numeric (Assuming 'sex' is a meaningful categorical feature)
# Check if 'sex' is in the categorical columns (it should be after convert_dtypes)
if 'sex' in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.barplot(x='sex', y='customer_age', data=merged_df)
    plt.title('Average Customer Age by Sex')
    plt.xlabel('Sex')
    plt.ylabel('Average Customer Age')
    plt.show()

    plt.figure(figsize=(8, 5))
    sns.barplot(x='sex', y='tenure', data=merged_df)
    plt.title('Average Tenure by Sex')
    plt.xlabel('Sex')
    plt.ylabel('Average Tenure')
    plt.show()

print("‚úÖ Visualizations generated.")
```
